```python
import json
from requests_oauthlib import OAuth1Session
from dateutil import parser
from pytz import timezone
import datetime
import csv
```

```python
API_KEY = "API Keyã®å€¤"
API_SECRET_KEY = "API Secret Keyã®å€¤"
ACCESS_TOKEN = "Access Tokenã®å€¤"
ACCESS_TOKEN_SECRET = "Access Token Secretã®å€¤"
```

### ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ãƒ¼ã®è¨­å®š

```python
tweet_count = 50
result_type = "recent"
search_keyword = "ã‚¨ã‚¢ãƒªã‚ºãƒ " # å•†å“åãªã©æ¤œç´¢ã—ãŸã„ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰ã‚’å…¥åŠ›
params = {"count": tweet_count, "result_type":result_type, "q":search_keyword}
```

### ãƒ„ã‚¤ãƒ¼ãƒˆãƒ‡ãƒ¼ã‚¿ã®å–å¾—

```python
url = "https://api.twitter.com/1.1/search/tweets.json"

twitter = OAuth1Session(API_KEY, API_SECRET_KEY, ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
response = twitter.get(url, params=params)
twitter_data = json.loads(response.text)
```

### ãƒ„ã‚¤ãƒ¼ãƒˆãƒ‡ãƒ¼ã‚¿ã®æŠ½å‡º

```python
tweet_list = []
for tweet in twitter_data["statuses"]:
    user_name = tweet["user"]["name"]
    followers_count = tweet["user"]["followers_count"]
    created_at = str(parser.parse(tweet["created_at"]).astimezone(timezone("Asia/Tokyo")))
    tweet_text = tweet["text"]
    tweet_list.append([user_name, followers_count, created_at, tweet_text])
```

```python
csv_header = ["ãƒ¦ãƒ¼ã‚¶ãƒ¼å","ãƒ•ã‚©ãƒ­ãƒ¯ãƒ¼æ•°","ä½œæˆæ—¥","ãƒ„ã‚¤ãƒ¼ãƒˆæ–‡"]
csv_date = datetime.datetime.today().strftime("%Y%m%d%H")
csv_file_name = "twitter_" + csv_date + ".csv"
with open(csv_file_name,"w", errors = "ignore") as file:
    writer = csv.writer(file,lineterminator="\n")
    writer.writerow(csv_header)
    writer.writerows(tweet_list)
```
